```python
input = pd.read_csv('/Users/samgould/Desktop/FSR Lab/finalized libraries/TP53 human/p53_combined_FINAL2.csv')
input
```

```python
ins = p53_filtered[p53_filtered['variant_type']=='INS']
dels = p53_filtered[p53_filtered['variant_type']=='DEL']
```

```python
aas = pd.read_csv('/Users/samgould/Desktop/FSR Lab/reference files/amino-acids.txt')
aa_dict = dict(zip(aas['three letter code'], aas['single letter code']))

three = aas['three letter code']

hg_ins = input[input['Variant_Type']=='INS']['HGVSp']

hg_trans = []
for i in hg_ins:
    sub = i[2:]
    for k in three:
        if k in sub:
            sub = sub.replace(k, aa_dict[k])

    hg_trans.append(sub)

patient_ins= []
other_ins = []
for i in np.unique(ins['HGVSp']):
    if i in hg_trans:
        patient_ins.append(i)
    else:
        other_ins.append(i)

hg_ins = input[input['Variant_Type']=='DEL']['HGVSp']

hg_trans = []
for i in hg_ins:
    sub = i[2:]
    for k in three:
        if k in sub:
            sub = sub.replace(k, aa_dict[k])

    hg_trans.append(sub)

patient_del= []
other_del = []
for i in np.unique(dels['HGVSp']):
    if i in hg_trans:
        patient_del.append(i)
    else:
        other_del.append(i)
```

```python
ins_dict = dict(zip(patient_ins, ['Observed indel']*len(patient_ins)))
ins_dict2 = dict(zip(other_ins, ['Random indel']*len(other_ins)))
ins_dict.update(ins_dict2)

del_dict = dict(zip(patient_del, ['Observed indel']*len(patient_del)))
del_dict2 = dict(zip(other_del, ['Random indel']*len(other_del)))
del_dict.update(del_dict2)

ins_dict.update(del_dict)
```

```python
#adding in annotations
new_no = []
classif = []

for i, val in p53_filtered.iterrows():

    hgvsp = val['HGVSp']
    var_type = val['variant_type']
    no = val['num_occurences']

    if var_type in ['INS', 'DEL']:
        cl = ins_dict[hgvsp]
        classif.append(cl)
        if cl == 'Observed indel':
            new_no.append(no)
        else:
            new_no.append(0)
    
    elif var_type == 'ONP':
        new_no.append(0)
        classif.append('Silent')

    else:
        new_no.append(no)
        classif.append('Observed SNV')



```

```python
p53_filtered['num_occurences'] = new_no
p53_filtered['classification'] = classif
```

```python
def domain_annotations(merged):

    TAD = [1,61]#=1 #TAD1/2
    PRR = [61,94]#=2 #PRR
    DBD = [94,292]#=3 #DBD
    NLS = [304,321]#=4 #NLS
    OD = [325,356]#=5 #OD
    CTD = [356,394]#=6 #CTD


    key_dict = {'TAD':TAD, 'PRR':PRR, 'DBD':DBD, 'NLS':NLS, 'OD':OD, 'CTD':CTD}
    keys1 = []
    for i, val in merged.iterrows():
        cod = val['codon']

        counter = 0
        for key in key_dict.keys():
            g1 =key_dict[key]
            lower = g1[0]
            upper = g1[1]

            if (cod>=lower) & (cod<upper):
                keys1.append(key)
                counter+=1
        
        if counter==0:
            keys1.append('Other')
        
        
    merged['Domain']=keys1

    return merged


p53_filtered= domain_annotations(p53_filtered)
```
